```python
def make_params(year_months):
    params = [ym.split('-') for ym in year_months]
    years = [int(ym[0]) for ym in params ]
    months = [int(ym[1]) for ym in params ]
    return years, months
```

```python
years, months = make_params(sales.month.values)
```

```python
sales['y'] = years
sales['m'] = months
```

```python
year_sales = sales.groupby('y').agg({'sales':'sum'})
```
